```python
import zipfile

def unzip_csv(zip_file_path, extract_to_folder):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        for file in zip_ref.namelist():
            if file.endswith('.csv'):
                zip_ref.extract(file, extract_to_folder)
                print(f'Extracted {file} to {extract_to_folder}')

# Example usage:
zip_file_path = '/content/India Agriculture Crop Production.csv.zip'
extract_to_folder = 'path/to/destination/folder'
unzip_csv(zip_file_path, extract_to_folder)
```

```python
#defining funtion for continues vs continues
from scipy import stats

def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(50):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.055:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False
```

```python
two_sample(df["Production"],df["Yield"])
```

```python
#defining function for categories vs categories
def chisqare_cat_vs_cat(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.055 else False
```

```python
chisqare_cat_vs_cat(df["State"],df["District"])
```

```python
#defining function for continous(d1) vs categories(d2)

def annova_test(d1,d2):
  group=df[d2].unique()
  data={}
  for i in group:
    data[i]=df[d1][df[d2]==i]
  f_value,p_value=stats.f_oneway(*[i for i in data.values()])  #task for baby
  if p_value < 0.055:
    return False
  else:
    return True
```

```python
annova_test("Production","State")
```

```python
final={}
for i in df.columns:
  final[i]={}
  for j in df.columns:
    if (i in continues) and (j in continues):
      result=two_sample(df[i],df[j])
    elif  (i in continues) and (j in categories):
      result=annova_test(i,j)
    elif (i in categories) and (j in continues):
      result=annova_test(j,i)
    elif (i in categories) and (j in categories):
      result=chisqare_cat_vs_cat(df[i],df[j])
    if result:
      final[i][j]=1
    else:
      final[i][j]=0
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
model1=DecisionTreeRegressor().fit(x_train,y_train)
y_pred1=model1.predict(x_test)
print(mean_squared_error(y_test,y_pred1))
print(mean_absolute_error(y_test,y_pred1))
print(r2_score(y_test,y_pred1))
```

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor(n_estimators=2,max_depth=4,random_state=10,min_samples_split=2).fit(x_train,y_train)
y_pred2=model2.predict(x_test)
print(mean_squared_error(y_test,y_pred2))
print(mean_absolute_error(y_test,y_pred2))
print(r2_score(y_test,y_pred2))
```
